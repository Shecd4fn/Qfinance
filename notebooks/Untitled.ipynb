```python
# Moving Average Crossover Strategy
# This notebook demonstrates how to use the `strategy.py` code to backtest a simple moving average crossover strategy.
```

```python
# Import necessary libraries and custom modules
import sys
sys.path.append('../src')  # This allows Python to find your strategy.py file in the src directory

from strategy import get_stock_data, calculate_moving_averages, generate_signals
import matplotlib.pyplot as plt
```

```python
# Load stock data for a specific ticker and date range
stock_data = get_stock_data('AAPL', '2022-01-01', '2023-01-01')

# Calculate moving averages
stock_data = calculate_moving_averages(stock_data)

# Generate buy/sell signals
stock_data = generate_signals(stock_data)
```

```python
# Visualize the stock data along with moving averages and signals
plt.figure(figsize=(14, 7))
plt.plot(stock_data['Close'], label='Close Price', alpha=0.5)
plt.plot(stock_data['SMA20'], label='20-day SMA', alpha=0.75)
plt.plot(stock_data['SMA50'], label='50-day SMA', alpha=0.75)
plt.scatter(stock_data.index, stock_data['Position'] == 1, color='green', label='Buy Signal', marker='^', alpha=1)
plt.scatter(stock_data.index, stock_data['Position'] == -1, color='red', label='Sell Signal', marker='v', alpha=1)
plt.title('Moving Average Crossover Strategy')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()
```
